```python
import pandas as pd
messages = pd.read_csv('Text_to_Vector\data\SMSSpamCollection.csv', sep='\t', names=["label", "message"])
```

```python
messages
```

```python
## Data cleaning and preprocessing:

import re
import nltk

nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer  ## use for bow
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]',' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review =' '.join(review)
    corpus.append(review)
```

```python
corpus
```

# Create bag of words

```python
# create bag of words model 
from sklearn.feature_extraction.text import CountVectorizer
# for binary bow enable binary=True
cv = CountVectorizer(max_features=2500,ngram_range= (1,2))
```

```python
# indpedent feature:
X = cv.fit_transform(corpus).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=30,linewidth=100000,
    formatter=dict(float=lambda x: "%.3g" % x))
```

```python
X
```

```python
cv.vocabulary_
```

```python
## output features

y=pd.get_dummies(messages['label'])

# Convert boolean values to integers (0 and 1)
y = y.astype(int)
```

```python
y
```

```python
y=y.iloc[:,0].values
```

```python
## train test split:

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.naive_bayes import MultinomialNB  ##  work well with sparse matrix data

spam_detect_model = MultinomialNB().fit(X_train, y_train)

```

```python
y_pred = spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report

accuracy_score(y_test, y_pred)  # compare results
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
